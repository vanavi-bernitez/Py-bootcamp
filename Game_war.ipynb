```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:

    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        #to call this magic method, use print(object instance)
        return f"{self.rank} of {self.suit}"
```

```python
from random import shuffle

class Deck:
    
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                #create the Card objects allows to use the __str__, so i cant print all the cards
                created_card = Card(suit, rank) 
                self.all_cards.append(created_card)

    def shuffle_cards(self):
        #don't return anything, becouse it work in place. 
        #I can't assign the result of the shuffle call in a variable
        shuffle(self.all_cards)

    #remove the last card from the list cos theres no specified index
    def remove_card(self):
        return self.all_cards.pop()

# new_card = Deck()

#takes the last card from the ordered deck
# last_card = new_card.all_cards[-1] 
# print(last_card)

#takes the last card from the shuffle deck
# new_card.shuffle_cards()
# last_card = new_card.all_cards[0:3]
# print(last_card[0], last_card[1], last_card[2])         
```

```python
class Player:
    def __init__(self,name):
        self.player_hand = []
        self.name = name

    def remove_player_card(self):
        #remove the top card from the player hand
        return self.player_hand.pop(0)

    def add_player_card(self, new_player_cards):
        #if new cards is a list, we use extend. Takes a list and merge into another.
        #Append in this case gets me a nested list. 
        #<if> statement means: if player grab a couple of cards 
        if type(new_player_cards) == type([]):
            self.player_hand.extend(new_player_cards)
        #if player grab ust one card 
        else:
            self.player_hand.append(new_player_cards)

    def __str__(self):
        return f"{self.name} has {len(self.player_hand)} cards"

# player1 = Player("pepo")
# player1.add_player_card(last_card)
# print(player1)
```
